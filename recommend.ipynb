```python
import numpy as np
import pandas as pd


movies = pd.read_csv('https://www.dropbox.com/s/vd8etoxpj9ctfko/tmdb_5000_movies.csv?dl=1')
credits = pd.read_csv('https://www.dropbox.com/s/5linnvejcaed39v/tmdb_5000_credits.csv?dl=1')

# movies = movies.merge(credits,on='title')
# movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
credits
```

```python

movies.dropna(inplace=True)  # remove the empty data rows. It is used to remove the data redundancy

import ast
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])

    return l


def convert_cast(obj):
    l = []
    count = 0
    for i in ast.literal_eval(obj):
        if (count != 3):
            l.append(i['name'])
            count += 1
        else:
            break
    return l


def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if (i['job'] == 'Director'):
            l.append(i['name'])
            break
    return l


# processing the genres coloumn
movies['genres'] = movies['genres'].apply(convert) #process the coloumn format into list
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x]) #removing the spaces between two words

# processing the keywords coloum
movies['keywords'] = movies['keywords'].apply(convert)
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x]) #removing the spaces between two words

# processing the cast coloum
movies['cast'] = movies['cast'].apply(convert_cast)
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x]) #removing the spaces between two words

# processing the crew coloum
movies['crew'] = movies['crew'].apply(fetch_director)
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x]) #removing the spaces between two words

#converting the overview coloumn values from string to a list and make different word
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']

# make the new dataframe having three coloum movie_id, title, tags
new_df = movies[['movie_id', 'title', 'tags']]

# convert tag coloumn values to string from list
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))

# convrt the tags coloumn value to lowercase
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())

new_df

# new_df.drop_duplicates(subset=['title'],inplace=True) # remove duplicates from the new_df dataframe
```

```python
# vectorization


import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python

vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similar = cosine_similarity(vectors)
similar
# similar is the matrix that contains cosine distances of each movie to other movies
# the more the distance the less the similarity between movies
```

```python
def recommend(movie):

    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similar[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)

recommend('Superman')        

```

```python
import pickle
pickle.dump(new_df.to_dict(),open('movies.pkl','wb')) # we have to transform the required data into a dictionary for sending them through pickle

pickle.dump(similar,open('similarity.pkl','wb')) # we need to send the similarity matrix too for fetching the results
```
